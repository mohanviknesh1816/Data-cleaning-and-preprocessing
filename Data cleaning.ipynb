```python
#CALLING DATASET

import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\91739\Downloads\Customer Call List.xlsx")
df
```

```python
#REMOVING DUPLICATES
df.drop_duplicates()
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("./_")
df
```

```python

df['Phone_Number'] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
# convering Phone_Numbervalues from float to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# inserting '-' between the phone numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df = df.drop(columns = "Phone_Number")
df
```

```python
# normalizing Paying Customer values
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

# normalizing Do_Not_Contact values
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')

df
```

```python
#numerical encoding
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', '1')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', '0')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', '1')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', '0')
df
```

```python
#filling Null values
df = df.fillna('')
df
```
